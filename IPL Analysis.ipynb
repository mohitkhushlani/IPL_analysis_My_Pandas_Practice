### isin() Function\

#### Jab kabhi b masking k liye bohot saari conditions bane tab mostly use hota h like tumko bohot saare masks create karne padh rahe h then isin function is ur savior

```python
## Lets find out the most destructive batsman (that is based on its strikerate) in the death over 
## batsman must have played 200 balls in 16-20 over 
```

```python
death_over=delivery[delivery['over']>15]
```

```python
death_over
```

```python
all_batsman=death_over.groupby('batsman')['batsman_runs'].count()
```

```python
x=all_batsman>200
```

```python
batsman_list=all_batsman[x].index.tolist()
```

```python
final=delivery[delivery['batsman'].isin(batsman_list)]
```

```python
runs=final.groupby("batsman")['batsman_runs'].sum()
```

```python
balls=final.groupby("batsman")['batsman_runs'].count()
```

```python
strike_rate=(runs/balls)*100
```

```python
strike_rate.sort_values(ascending=False).head()
```
